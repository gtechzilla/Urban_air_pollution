```python
# dropping columns not in our test data
for i in train.columns:
  if i not in test.columns:
    print(i)

target = train.target
target.head()

```

```python
# splitting our date year and time columns
# Train set
train['Year'] = pd.to_datetime(train['Date']).dt.year
train['Month'] = pd.to_datetime(train['Date']).dt.month
train['Day'] = pd.to_datetime(train['Date']).dt.day

# Test data
test['Year'] = pd.to_datetime(test['Date']).dt.year
test['Month'] = pd.to_datetime(test['Date']).dt.month
test['Day'] = pd.to_datetime(test['Date']).dt.day
```

```python
# Filling the present null values in the train dataset
train = train.fillna(train.mean())
train.info()
```

```python
train.isnull().sum().any()
train.isnull().sum()
```

```python
# Splitting our data
X= train.drop(["Place_ID X Date", "Date", "Place_ID", "target", "target_min", "target_max", "target_variance", "target_count"], axis=1)
Y = train.target
```

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, Y_train, Y_valid = train_test_split(X, Y, test_size=0.3, random_state=1) 
# X_train.shape
Y_train.shape
```

```python
# Modelling using CatBoostRegressor
from catboost import CatBoostRegressor

predictions = []
# predicting using different seeds
for i in range(10):
  model = CatBoostRegressor(verbose=False,
                            depth=9,
                            iterations=1500,
                            random_seed=i)
  model.fit(X, Y)

# making and averaging the predictions
test.drop(["Place_ID X Date", "Date", "Place_ID"], axis=1, inplace=True)
preds=model.predict(test)
predictions.append(preds)
# averaging
predss = np.mean(predictions, axis=0)


```
